# Loading data to variable

```
dataset.shape

dataset = dataset.dropna() #dropping rows with no data
dataset.shape
dataset
y = dataset.iloc[:, -1].values
X = dataset.iloc[:,[1,3]] # storing column 1 and 3 into X
print(len(dataset))
print(dataset['title'].shape)
print(dataset['text'].shape)
#X.reset_index(inplace=True)
y = np.reshape(y,(18285, 1))
y.shape
```

# Cleaning the text

```
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
title1 = []
text1 = []
for i,v in enumerate(X['title']):
    titl = re.sub('[^a-zA-Z]', ' ', v) #Taking only text 'a-z & A-Z' and replacing other things with a space
    titl = titl.lower() #converting the text into lowercase
    titl = titl.split() #splitting to get a list of words
    ps = PorterStemmer() #instanciating PorterStemmer for Stemming
    titl = [ps.stem(word) for word in titl if not word in set(stopwords.words('english'))] # removing useless words in english eg: i, me, is, am while containing the meaning
    titl = ' '.join(titl) # Joining the semmed words into a single line
    title1.append(titl)

#for i,v in enumerate(X['text']):
#    text = re.sub('[^a-zA-Z]', ' ', v)
#    text = text.lower()
#    text = text.split()
#    ps = PorterStemmer()
#    text = [ps.stem(word) for word in text if not word in set(stopwords.words('english'))]
#    text = ' '.join(text)
#    text1.append(text)

#text
```

# Creating Bag of Word Model

```
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer() #Instanciating CountVectorizer
x = cv.fit_transform(title1).toarray() # taking each word converting their occurance into an array
y = dataset.iloc[:, -1].values # loading the labels
print(x.shape)
```

# Splitting dataset into training set and test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0) #splitting the dataset into 80% data for training and 20% for testing
```

# Training SVM model

```
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train,y_train)
```

# Training KNN model

```
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_train, y_train)
```

# Predicting for KNN and forming confusion matrix

```
y_pred=neigh.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Predicting for SVM and forming confusion matrix

```
y_pred=clf.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Training model for Random FOrest

```
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state=0)
clf.fit(X_train,y_train)
```

# Predicting for Random Forest and forming confusion matrix

```
y_pred=clf.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
